```python
#Reference : https://towardsdatascience.com/extract-features-of-music-75a3f9bc265d
import librosa
import librosa.display
import matplotlib.pyplot as plt
def calculate_fft(filename,dr,count,traindirectory):

    waveform,sr=librosa.load(filename)
    #print("Sampling Rate: "+str(sr))
    #waveform is a 1 dimensional numpy array that stores the values of an audio signal wrt time
    #sr is the sampling rate of the audio signal
    frame_size = int(0.5 * sr)
    #print("Frame Size: "+str(frame_size))
    #Each 10 second audio is divided into frames each of 0.5 seconds each
    stride = int(0.1 * sr)
    #print("Stride: "+str(stride))
    #Here stride is taken to be 0.1 seconds
    mfccs=librosa.feature.mfcc(waveform,sr=sr,n_mfcc=13,hop_length=stride,n_fft=frame_size)
    #Here, number of features = 13
    
    #print(mfccs.shape)
    img = librosa.display.specshow(mfccs,sr=sr,x_axis='time')
    #print(type(img))
    
    img.figure.savefig(train_directory+dr+"/"+str(count1))
    #fig = plot.get_figure()
    #fig.savefig(str(count1)+".jpg")
    
    return mfccs,img
    
    
    #print(type(waveform),type(sr))
    
    
    
    
    
    
```

```python
import cv2

categories=[]
for category in list_of_files.keys():
    categories.append(category)
    
def get_index(category,categories):
    print(category)
    for i in range(0,len(categories)):
        if(categories[i]==category):
            break
    return(i+1)
    






    
                
                
          

                
                
        
        
    
    
```

```python
import random

list1=[x for x in range(100)]
print("List1:")
print(list1)

training_data=[]
test_data=[]

test_index=[]
directory = '/kaggle/working/train/'
for category in list_of_files.keys():
    category1 = category + "/"
    path = os.path.join(directory,category1)
    print(path)
    class_label = get_index(category,categories)
    for direc,_,filenames in os.walk(path):
          print("Length of filenames"+str(len(filenames)))
          print(filenames)
          train_index = random.choices(list1,k=70)
          for i in list1:
                if i not in train_index:
                    test_index.append(i)
          
          for i in train_index:
                full_path=os.path.join(path,filenames[i])
                img_arr = cv2.imread(full_path) 
                #print(img_arr.shape)
                im_rgb = cv2.cvtColor(img_arr,cv2.COLOR_BGR2RGB)
                print(im_rgb.shape)
                #plt.imshow(im_rgb)
                #plt.show()
                training_data.append([im_rgb,class_label])
                
                
          for i in test_index:
                full_path=os.path.join(path,filenames[i])
                img_arr = cv2.imread(full_path) 
                #print(img_arr.shape)
                im_rgb = cv2.cvtColor(img_arr,cv2.COLOR_BGR2RGB)
                print(im_rgb.shape)
                #plt.imshow(im_rgb)
                #plt.show()
                test_data.append([im_rgb,class_label])
                
          
    
            
                
#print(len(training_data))
```
