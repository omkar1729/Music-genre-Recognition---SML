```python
import cv2
def get_index(category,categories):
    print(category)
    for i in range(0,len(categories)):
        if(categories[i]==category):
            break
    return(i)

def genTrainTest(training_data,test_data,list_of_files,categories,directory1,directory2):
    list1 = [x for x in range(100)] #Since we have 100 files in each genre, I take a list which contains values like [0,1,2,3,4,.....,100].
    list2 = list1 #create a backup of list1
    for category in list_of_files.keys(): #For each music genre
        category1 = category + "/"
        path1 = os.path.join(directory1,category1) #helps to navigate through the required genre directory
        path2 = os.path.join(directory2,category1)
        #print(path)
        class_label = get_index(category,categories) #we have to represent each category by a number which is nothing but the position of the genre name in the categories list. 
        test_index=[]
        for direc,_,filenames in os.walk(path1):
              #print("Length of filenames"+str(len(filenames)))
             # print(filenames)
              random.shuffle(list1) #shuffling the list 
              train_index = list1[:90] #take the first 70 random indices as train indices
              #print("List1: "+str(list1))
              #print("Training index: "+str(train_index))
              #print(len(train_index))
              #count=0
              #count1 = 0
              #count2=0
              #print(len(list1))
              for i in range(0,len(list1)):
                if(list1[i] not in train_index):
                    test_index.append(list1[i]) #remaining 30 for test indices
              #print(count,count1,count2)
              #print(train1_index==train_index)
              #test_index = list(set(list1)-set(train_index))
              #print(len(test_index))
              #print("Test index: "+str(test_index))


              for i in train_index:
                    full_path1=os.path.join(path1,filenames[i]) #navigate to each genre subfolder
                    full_path2=os.path.join(path2,filenames[i])
                    img_arr1 = cv2.imread(full_path1) #Read the image
                    img_arr2 = cv2.imread(full_path2) #Read the image
                    #print(img_arr)
                    im_rgb1 = cv2.cvtColor(img_arr1,cv2.COLOR_BGR2RGB) #Convert into RGB as colour is important
                    im_rgb2 = cv2.cvtColor(img_arr2,cv2.COLOR_BGR2RGB) 
                    #print(im_rgb.shape)


                    #plt.imshow(im_rgb)
                    #plt.show()
                    training_data.append([im_rgb1,im_rgb2,class_label]) #each row in the training data contains the rgb numpy matrix and the class label 



              for i in test_index:
                    full_path1=os.path.join(path1,filenames[i])
                    full_path2=os.path.join(path2,filenames[i])
                    img_arr1 = cv2.imread(full_path1)
                    img_arr2 = cv2.imread(full_path2)
                    #print(img_arr.shape)
                    im_rgb1 = cv2.cvtColor(img_arr1,cv2.COLOR_BGR2RGB)
                    im_rgb2 = cv2.cvtColor(img_arr2,cv2.COLOR_BGR2RGB)
                    #print(im_rgb.shape)
                    #plt.imshow(im_rgb)
                    #plt.show()
                    test_data.append([im_rgb1, im_rgb2 ,class_label]) #similarly for test data
                    
                    
    return training_data,test_data
```

```python
def convert_to_numpy(training_data,test_data):
    X1=[]
    X2=[]
    Y=[]
    for features1,features2,class_label in training_data:
        X1.append(features1)
        X2.append(features2)
        Y.append(class_label)
        
    #X has to be a numpy array
    Y=np.asarray(Y)
    #X=np.asarray(X)
    #print(len(X[0]))
    #print(len(X[0][0]))
    #print(len(X[0][0][0]))
    #X=np.array(X)
    X1=np.asarray(X1)
    X2=np.asarray(X2)
    #print(X.shape)
    #print(Y.shape)
    np.save("XTrain1.npy",X1)
    np.save("XTrain2.npy",X2)
    np.save("YTrain.npy",Y)


    X21 = []
    X22=[]
    Y2 = []
    for features1,features2, class_label in test_data:
        X21.append(features1)
        X22.append(features2)
        Y2.append(class_label)
    X21 = np.asarray(X21)
    X22 = np.asarray(X22)
    Y2 = np.asarray(Y2)
    np.save("XTest1.npy",X21)
    np.save("XTest2.npy",X22)
    np.save("YTest.npy",Y2)


    return(X1,X2,Y,X21,X22,Y2)
```
